```python
import numpy as np
import pandas as pd
import matplotlib as mt
import seaborn as sns
```

```python
df = pd.read_csv("medical_examination.csv")
```

```python
df
```

```python
df["BMI"] = df["weight"]/(df["height"]/100)**2
```

```python
df["BMI"]
```

```python
df
```

```python
df.loc[df["BMI"]>25,"overweight"] = 1
df.loc[df["BMI"]<25,"overweight"] = 0
```

```python
df
```

```python
df["cholesterol"].min()|df["gluc"].min()
```

```python
df["cholesterol"].max()|df["gluc"].max()
```

```python
df['normalized_value'] = ((0.99 - 0.01) * (df['cholesterol']|df["gluc"] -df["cholesterol"].min()|df["gluc"].min() ) / (df["cholesterol"].max()|df["gluc"].max() -df["cholesterol"].min()|df["gluc"].min() )) + 0.01
```

```python
df["normalized_value"]
```

```python
df
```

```python
df.loc[df["normalized_value"]>0.5,"condition"]="Good"
df.loc[df["normalized_value"]<0.5,"condition"]="Bad"
```

```python
df
```

```python
A = pd.melt(df,id_vars="cardio",value_vars=["cholesterol","gluc","smoke","alco","active","overweight"])
```

```python
A
```

```python
A["total"] = 1
A = A.groupby(["cardio","variable","value"],as_index=False).count()
```

```python
ax = sns.catplot(data=A,x="cardio",y="total",hue="value",kind="bar",col="cardio")
```
